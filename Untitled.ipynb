```python
import pandas as pd

from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
output_notebook()
```

```python
zips = pd.read_csv('./Zip_zhvi_bdrmcnt_3_uc_sfrcondo_tier_0.33_0.67_sm_sa_mon.csv')
```

```python
port_zips = zips.loc[
    (zips['City'].str.contains('Portland')) &
    (zips['State'].str.contains('OR'))
]

# port_zips = zips.loc[
#     (zips['RegionName']==28209)
# ]
```

```python
for col in ['RegionID',
 'SizeRank',
 'RegionType',
 'StateName',
 'State',
 'City',
 'Metro',
 'CountyName',]:
    del port_zips[col]
```

```python
port_zips = port_zips.set_index('RegionName').transpose()

port_zips.index = pd.to_datetime(port_zips.index)

dates = list(port_zips.index)

```

```python
def plotme(df):
    TOOLTIPS = [
        ("zip", "$name"),
    ]

    p = figure(plot_width=800, plot_height=400, x_axis_type='datetime', tooltips=TOOLTIPS,)

    # data = ColumnDataSource(port_zips)
    for zip_code in df.columns:
        y = list(df[zip_code])
        p.line(x=dates, y=y, name=str(zip_code))


    show(p)
plotme(port_zips)
```

```python
port_zips.head()
```

```python
growth = port_zips.pct_change() * 12 * 100
```

```python
growth
```

```python
# annual pct change each month by zip code
plotme(growth)
```

```python
# pct change climbing or falling?
plotme(growth / abs(growth))
```
